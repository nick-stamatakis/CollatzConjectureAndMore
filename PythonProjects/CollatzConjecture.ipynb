```python
#Asks whether repeating two simple arithmetic operations will eventually transform every positive integer into 1
#Known as the Collatz Conjecture
def collatz(num):
    while num != 1:
        print(num)
        #if even divide by 2
        if num % 2 == 0:
            num = int(num / 2)
        #if odd multiply by 3 and add 1
        else:
            num = int(3 * num + 1)
    else:
        print("We were able to transform the number into")
        print(num)

#Main method to test my code and take in user input
def main():
    num = int(input("Please input an integer >= 1: "))
    collatz(num)
    
#Call main function    
main()
```
